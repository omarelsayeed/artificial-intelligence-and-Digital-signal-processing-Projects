```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style
%matplotlib notebook
```

```python
#create signals
t=np.arange(0,1,0.001)
n=len(t)
f=np.sin(2*np.pi* 50*t) + np.sin(2*np.pi*120*t)
f_clean=f
f=f+ 2.5*np.random.randn(len(t))
```

```python
#FFT
freq_comp=np.fft.fft(f)
freq_amp=np.absolute(freq_comp)[0:int(len(freq_comp)/2)]

x_axis=(1/(0.001*n) ) *np.arange(n)
x_axis=x_axis[0:int(n/2)]

matplotlib.style.use('dark_background')
plt.figure(figsize=(15,7))
plt.xlabel('Freq (Hz)')
plt.ylabel('Amplitude')
plt.plot(x_axis,freq_amp , color='green')
```

```python
#Spectogram
plt.specgram(f , NFFT=128 , Fs=1000 , noverlap=120 , cmap='jet_r')
plt.colorbar()
plt.show()
```

```python
#Filtering with the Fourier transform
for i in range(len(freq_amp)):
    if freq_amp[i]<200:
        freq_amp[i]=0
plt.plot(freq_amp)
```
