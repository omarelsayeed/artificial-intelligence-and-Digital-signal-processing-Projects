```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from matplotlib import style
from scipy.interpolate import UnivariateSpline , Rbf
```

```python
x=np.linspace(0,4,12)
y=np.cos(x**2/3+4)
```

```python
plt.plot(x,y , 'bo')
```

```python
f1=interp1d(x,y , kind='linear') #returns a function
f2=interp1d(x,y, kind='cubic') #returns a function
```

```python
x_test=np.linspace(0,4,30)
plt.plot(x_test,f1(x_test))
plt.plot(x_test,f1(x_test) , 'bo')
```

```python
x_test=np.linspace(0,4,30)
plt.plot(x_test,f2(x_test))
plt.plot(x_test,f2(x_test) , 'bo')
```

```python
#You can use interpolation to reconstruct signals
#You can use it as a regression model
```

```python
#Create random function for x in range of -3 , 3
x=np.linspace(-3,3,50)
y=np.exp(-x**2) + 0.1*np.random.randn(50)
plt.plot(x,y , 'bo')
```

```python
#Spline interpolation
spl=UnivariateSpline(x,y)
xs=np.linspace(-3,3,1000)
spl.set_smoothing_factor(0.5)
plt.plot(x,y , 'bo')
plt.plot(xs,spl(xs))
```

```python
#rbf interpolation "radial basis function"
x=np.linspace(0,10,9)
y=np.sin(x)
plt.plot(x,y,'ro')
plt.plot(x,y)
```

```python
x_test=np.linspace(0,10,101)
rbff=Rbf(x,y)
plt.plot(x_test , rbff(x_test))
plt.plot(x_test , rbff(x_test) , 'ro')
#it learnt that the function is a sin wave from only 9 DATA POINTS !
```
