```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create A Signal
freq=50
t_period=1/freq #20 ms
time=t_period * 2 # go for 40ms
amplitude = 2

t=np.linspace(0,time, 500 , endpoint=True) #sample rate = 500
signal=amplitude * np.sin(2 * np.pi * freq * t)

#Add noise if you want
#signal = signal + np.random.rand(len(signal)) #with noise
```

```python
#Ws>2Wm
Fsampling=1000 #Sampling Frequency
ts=1/Fsampling 
txs=np.arange(0, (time+(ts/2)) , ts) 
x=2*3.14*freq*t
r=np.round(len(t)/len(txs)) 
xts=np.arange(0,len(t) , r).astype('int')
xs=x[xts] 
ys=signal[xts]
```

```python
axes=plt.subplot()
axes.plot(xs,ys , linewidth=3,linestyle='' , marker='o')
axes.plot(x,signal)
axes.bar(xs,ys , bottom=0,width=0.1 , color='blue')
axes.set_xticks((2*3.14*freq) * (np.arange(0,41,5)*1e-3))
axes.set_xticklabels(np.arange(0,41,5) , fontsize=14)
plt.show()
```

```python
#Sampling at the Nyquist Freq

Fsampling=100
ts=1/Fsampling 
txs=np.arange(0, (time+(ts/2)) , ts) 
x=2*3.14*freq*t
r=np.round(len(t)/len(txs)) 
xts=np.arange(0,len(t) , r).astype('int')
xs=x[xts] 
ys=signal[xts]
axes=plt.subplot()
axes.plot(xs,ys , linewidth=3,linestyle='' , marker='o')
axes.plot(x,signal)
axes.bar(xs,ys , bottom=0,width=0.1 , color='blue')
axes.set_xticks((2*3.14*freq) * (np.arange(0,41,5)*1e-3))
axes.set_xticklabels(np.arange(0,41,5) , fontsize=14)
plt.show()
```
